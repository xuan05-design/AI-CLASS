```python
import pandas as pd

# The previous operations failed because the 'df' DataFrame did not contain the expected columns
# for the Titanic dataset. The current 'df' (from kernel state) appears to be from a different dataset.
# To fix this, we will first load the Titanic dataset into 'df' from a public URL.
# If you intend to load 'df' from your Google Drive (e.g., from 'titanic.csv'),
# please ensure the loading step in an earlier cell (like 8c726b72) is corrected and executed first.
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

# Addressing FutureWarning by reassigning the results instead of using inplace=True
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
X = df.drop('Survived', axis=1)
y = df['Survived']

print(df.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
```
